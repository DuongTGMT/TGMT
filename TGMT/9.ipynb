```python
import matplotlib as mpl
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
```

```python
my_path = 'C:/Users/HP/OneDrive/Tài liệu/TGMT/img01.jpg'
im = Image.open(my_path)
plt.imshow(im)
```

```python
im_gray = im.convert('L')
plt.imshow(im_gray, cmap='gray')
```

```python
im_array = np.array(im_gray)
print(im_array)
```

```python
plt.figure()
plt.hist(im_array.flatten(), bins=130)
plt.title('Biểu đồ Histogram')
plt.xlabel('pixel')
plt.ylabel('Số lượng')
plt.show()
```

```python
plt.figure()
plt.gray()
plt.contour(im_gray, origin = 'image')
plt.show()
```

```python
im = Image.open(my_path).convert('L')
plt.imshow(im)
plt.figure()
plt.gray()
plt.contour(im, origin = 'image')
plt.axis('equal')
plt.show()
```
