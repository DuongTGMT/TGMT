```python
from PIL import Image
```

```python
dir = 'C:/Users/HP/OneDrive/Tài liệu/TGMT/img.png'
```

```python
print(dir)
```

```python
img = Image.open(dir)
```

```python
display(img)
```

```python
region = (100, 75, 175, 160)
cropped_image = img.crop(region)
display(cropped_image)
```

```python
paste_position = (150, 150)
img.paste(cropped_image, paste_position)
display(img)
```

```python
print(img.size)
```

```python
new_size = (1000,1000)
new_image = img.resize(new_size)
display(new_image)
```

```python
thumbnail_size = (100,100)
img.thumbnail(thumbnail_size)
display(img)
```

```python
rotated_image = img.rotate(90)
display(rotated_image)
```
