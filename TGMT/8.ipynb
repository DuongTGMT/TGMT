```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [20, 10, 25, 16, 30]

x1 = [1, 2, 3, 4, 5]
y1 = [30, 15, 5, 10, 20]

plt.plot(x, y, 'bo-', label='Dữ liệu biểu đồ')
plt.plot(x1, y1, marker='o', color='r', linestyle='-', label='Dữ liệu biểu đồ-1')
plt.legend()
plt.xlabel('Trục x') 
plt.ylabel('Trục y') 
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D', 'E']
values = [5, 15, 25, 20, 18]
plt.bar(categories, values, color='g', alpha=0.5)
plt.xlabel('Danh mục')
plt.ylabel('Giá trị')
plt.show()
```

```python
my_labels = ['A', 'B', 'C', 'D']
sizes = [20, 50, 20, 10]
my_colors = ['red', 'green', 'blue', 'yellow']

plt.pie(sizes, labels=my_labels, colors = my_colors, autopct='%1.1f%%')
plt.title('Biểu đồ tròn')
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [20, 16, 19, 30, 20]
plt.scatter(x, y, c='r', marker='o', label='Nguồn')
plt.title('Biểu đồ phân tán')
plt.xlabel('Trục x')
plt.ylabel('Trục y')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
data = [20, 18, 19, 22, 30, 45, 35, 55, 65]
plt.boxplot(data)
plt.title('Biểu đồ hộp')
plt.ylabel('Giá trị')
plt.show()
```

```python
import matplotlib.pyplot as plt

data = [20, 18, 19, 30, 25, 45, 50, 30, 65]

plt.violinplot(data) 
plt.title('Biểu đồ violin')  
plt.ylabel('Giá trị') 
plt.show()
```
